```python
#definimos las primeras constantes
a=0.24250109
b=-0.0607718825
c=0.93240737
d=0.86473146
G=1
#tiempo inicializado en cero
t_inicial=0.0
m=1
#definimos los vectores
r1_0 = np.array([a,b])
r2_0 = np.array([-a,-b])
r3_0 = np.array([0.,0.])
v1_0 = np.array([c,d])
v2_0 = np.array([c,d])
v3_0 = np.array([-2*c,-2*d])
#condiciones iniciales
ini1=np.array([r1_0[0], r1_0[1], v1_0[0], v1_0[1]])
ini2=np.array([r2_0[0], r2_0[1], v2_0[0], v2_0[1]])
ini3=np.array([r3_0[0], r3_0[1], v3_0[0], v3_0[1]])
```

```python
#Funci[on para primera derivada
def sqrt_f(r):
    return np.sqrt(r[0]**2 + r[1]**2)

#Funcion para la derivada de un vector dado
def f(h1,h2,h3):
    #Posiciones de cada cuerpo
    r1 = np.array([h1[0],h1[1]])
    r2 = np.array([h2[0],h2[1]])
    r3 = np.array([h3[0],h3[1]])
    r_12 = r2 - r1 
    r_13 = r3 - r1
    r_23 = r3 - r2
    sqvecr12 = sqrt_f(r_12) 
    sqvecr13 = sqrt_f(r_13)
    sqvecr23 = sqrt_f(r_23)
    
    fx1 = m*(r_12[0]/(sqvecr12**3) + r_13[0]/(sqvecr13**3))
    fx2 = m*(r_12[1]/(sqvecr12**3) + r_13[1]/(sqvecr13**3))
    fx3 =-m*(r_12[0]/(sqvecr12**3) + r_23[0]/(sqvecr23**3))
    
    fy1 = -m*(r_12[1]/(sqvecr12**3) + r_23[1]/(sqvecr23**3))
    fy2 = -m*(r_13[0]/(sqvecr13**3) - r_23[0]/(sqvecr23**3))
    fy3 = -m*(r_13[1]/(sqvecr13**3) - r_23[1]/(sqvecr23**3))
    
    return np.array([h1[2],h1[3],fx1,fy1])
    return np.array([h2[2],h2[3],fx2,fy2])
    return np.array([h3[2],h3[3],fx3,fy3])
```

```python
def one_Kutta4_step():
    global t
    global Deltat
    global h1,h2,h3
    K1 = f(h1, h2, h3)
    K2 = f(h1 + Deltat*(1./2.*K1[0]), h2 + Deltat*(1./2.*K1[1]), h3 + Deltat*(1./2.*K1[2]))
    K3 = f(h1 + Deltat*(1./2.*K2[0]), h2 + Deltat*(1./2.*K2[1]), h3 + Deltat*(1./2.*K2[2]))
    K4 = f(h1 + Deltat*K3[0], h2 + Deltat*K3[1], h3 + Deltat*K3[2])
    
    h1+= Deltat * (1/6.*K1[0] + 1/3. * K2[0] + 1/3.*K3[0] + 1/6.*K4[0])
    h2+= Deltat * (1/6.*K1[1] + 1/3. * K2[1] + 1/3.*K3[1] + 1/6.*K4[1])
    h3+= Deltat * (1/6.*K1[2] + 1/3. * K2[2] + 1/3.*K3[2] + 1/6.*K4[2])
    t+= Deltat
    
#Con. iniciales
h1=ini1
h2=ini2
h3=ini3
```

```python
t=0.
iteraciones=5000
t_max=0.5
Deltat=t_max/iteraciones
```

```python
history1=np.array([[t_inicial,h1[0],h1[1],h1[2],h1[3]]])
history2=np.array([[t_inicial,h2[0],h2[1],h2[2],h2[3]]])
history3=np.array([[t_inicial,h3[0],h3[1],h3[2],h3[3]]])
```

```python
for __ in range(iteraciones):    
    one_Kutta4_step()
    history1=np.append(history1,[[t_inicial,h1[0],h1[1],h1[2],h1[3]]],axis=0)
    history2=np.append(history2,[[t_inicial,h2[0],h2[1],h2[2],h2[3]]],axis=0)
    history3=np.append(history3,[[t_inicial,h3[0],h3[1],h3[2],h3[3]]],axis=0)
```
